```python
import numpy as np
import matplotlib.pyplot as plt
import os
import librosa
import IPython.display as ipd
```

```python
BASE_DIR="C:/Users/User/EDD_audio/EATD-Corpus/t_1"
neu_file="neutral_out.wav"
pos_file="positive_out.wav"
neg_file="negative_out.wav"
neu_dir=(os.path.join(BASE_DIR, neu_file)).replace('\\','/')
pos_dir=(os.path.join(BASE_DIR, pos_file)).replace('\\','/')
neg_dir=(os.path.join(BASE_DIR, neg_file)).replace('\\','/')
print(neg_dir)
```

```python
ipd.Audio(neu_dir)
neu_lw, sr= librosa.load(neu_dir)
pos_lw, _= librosa.load(pos_dir)
neg_lw, _= librosa.load(neg_dir)
```

```python
print(neu_lw)
print(neu_lw.shape)
```

```python
def plot_mag_spec(signal, title, sr):
    ft= np.fft.fft(signal)
    mag_spec=np.abs(ft)
    plt.figure(figsize=(18,5))
    freq= np.linspace(0, sr, len(mag_spec))
    plt.plot(freq, mag_spec)
    plt.xlabel("Freq (Hz)")
    plt.title(title)
    plt.show()
```

```python
plot_mag_spec(neu_lw,"Neutral", sr)
```

```python
plot_mag_spec(pos_lw,"Positive", sr)
```

```python
plot_mag_spec(neg_lw,"Negative", sr)
```
